```python
# Import necessary PyTorch libraries
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.utils.data as data
import torchvision
import torchvision.transforms as transforms


# Additional libraries for visualization and utilitie
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_device():
    """Selects the best available device for PyTorch computations.

    Returns:
        torch.device: The selected device.
    """

    if torch.cuda.is_available():
        return torch.device('cuda')
    elif torch.backends.mps.is_available():
        return torch.device('mps')
    else:
        return torch.device('cpu')

device = get_device()
print(f"using device: {device}")
```

# Experiments on the MNIST dataset

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# Define the transformation with resizing
transform = transforms.Compose([
    transforms.Resize((28, 28)),  # Resize to 28x28 (not necessary for MNIST, but included for consistency)
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # Normalize with MNIST mean and std
])

# Load the MNIST dataset
dataset = datasets.MNIST(root='./data', download=True, transform=transform, train=True)

# Print the total number of images in the dataset
print(f"Total number of images in the dataset: {len(dataset)}")

# Splitting dataset into training and validation sets
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# Print the number of images in the train and validation sets
print(f"Number of images in the training set: {len(train_dataset)}")
print(f"Number of images in the validation set: {len(val_dataset)}")

# Create DataLoader instances
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False)

# Inspect the contents of the train_loader
train_batches = 0
for batch in train_loader:
    train_batches += 1

print(f"Number of batches in the training loader: {train_batches}")
print(f"Total number of images in the training loader: {train_batches * 128}")
```

## Training Samples

```python
# Function to denormalize the image
def denormalize(tensor, mean, std):
    tensor = tensor.clone()  # Clone the tensor to avoid modifying the original
    for t, m, s in zip(tensor, mean, std):
        t.mul_(s).add_(m)
    return tensor

# Get a batch of training data
dataiter = iter(train_loader)
images, labels = next(dataiter)

# Select 25 random images
images = images[:25]

# Denormalize the selected images
mean = [0.1307]
std = [0.3081]
images = denormalize(images, mean, std)

# Convert to numpy for plotting
images = images.numpy()

# Plotting the images
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
axes = axes.flatten()

for img, ax in zip(images, axes):
    img = np.transpose(img, (1, 2, 0))  # Convert from (C, H, W) to (H, W, C)
    ax.imshow(img.squeeze(), cmap='gray')  # Plot in grayscale
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Training MNIST with DDPM

```python
import torch
import os

def get_trained_epochs(filename="mnist_gaussian.pth"):
    """Loads the checkpoint and returns the number of epochs the model was trained for."""
    if os.path.isfile(filename):
        checkpoint = torch.load(filename, map_location=torch.device('cpu'))
        epoch = checkpoint['epoch']
        return epoch
    else:
        print(f"No checkpoint found at {filename}.")
        return -1

# Filename
filename = "mnist_gaussian.pth"
# Assuming the .pth file is present in the environment
end_epoch = get_trained_epochs(filename)
print(f"The model was trained for {end_epoch+1} epochs.")
```

```python
import torch
import matplotlib.pyplot as plt
import random

# Load the sampled images
sampled_data = torch.load("mnist_gaussian_ddpm.pt")

# Extract the images
images = [data['sampled'] for data in sampled_data.values()]

# Select 25 random images
selected_images = random.sample(images, 25)

# Convert the images to a suitable format for plotting
selected_images = torch.stack(selected_images).numpy()

# Plotting the images
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
axes = axes.flatten()

for img, ax in zip(selected_images, axes):
    img = img.squeeze()  # Remove the channel dimension if it's there
    ax.imshow(img, cmap='gray')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Sampled images and original images using algorithm 1

```python
import torch
import matplotlib.pyplot as plt
import random

# Load the sampled images and their corresponding original images
sampled_data = torch.load("mnist_echo_cold_l1_alg1.pt")

# Extract the keys (indices) and shuffle them to pick random samples
keys = list(sampled_data.keys())
random.shuffle(keys)

# Select 5 random keys
selected_keys = keys[:5]

# Extract the original and sampled images
original_images = [sampled_data[key]['original_image'] for key in selected_keys]
sampled_images = [sampled_data[key]['sampled'] for key in selected_keys]

# Convert images to numpy for plotting
original_images = torch.stack(original_images).numpy()
sampled_images = torch.stack(sampled_images).numpy()

# Plotting the images
fig, axes = plt.subplots(5, 2, figsize=(10, 25))

for i in range(5):
    # Plot original image
    ax = axes[i, 0]
    img = original_images[i].squeeze()
    ax.imshow(img, cmap='gray')
    ax.set_title("Original Image")
    ax.axis('off')
    
    # Plot sampled image
    ax = axes[i, 1]
    img = sampled_images[i].squeeze()
    ax.imshow(img, cmap='gray')
    ax.set_title("Sampled Image")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
